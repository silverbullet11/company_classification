```python
x_values = [1, 2, 3, 4, 5]
y_values = [x**2 for x in x_values]
y1_values = [x*2 for x in x_values]
colors = ['red', 'red', 'black', 'blue', 'green']

p = figure()
p.line(x=x_values, y=y_values, line_width=2)
p.circle(x=x_values, y=y_values, color='red', size=5, alpha=0.5)
show(p)
```

```python
data = {'x_values': x_values, 'y_values': y_values, 'y1_values': y1_values, 'colors': colors}
df = pd.DataFrame.from_dict(data)
df
```

```python
source = ColumnDataSource(data=df)
p = figure()
p.circle(x='x_values', y='y_values', color='colors', source=source)
```

```python
handler = show(p, notebook_handle=True)
```

```python
df.loc[5] = [6, 36, 12, 'red']
source.stream(df)
push_notebook(handle=handler)
```

```python
select = Select(title='Option:', value='1', options=[str(x) for x in x_values])
s2 = ColumnDataSource(data=dict(x=x_values, y=y_values))
callback = CustomJS(args=dict(s2=s2), code="""
    var inds = cb_obj.value;
    console.log(inds);
    var d2 = s2.data;
    console.log(d2);
    x = d2['x']
    y = d2['y'];
    for(var i=0; i<x.length; i++) {
        y[i] = x[i] * inds;
    }
    s2.change.emit();
""")
select.js_on_change('value', callback)

TOOLTIPS = [
    ('(x, y)', '($x, $y)'),
    ('(x1, y1)', '(@x, @y)')
]

interact_plot = figure(plot_width=300, plot_height=300, title='Interact Plot', tools=tools)
interact_plot.circle('x', 'y', source=s2)
hover = interact_plot.select(dict(type=HoverTool))
hover.tooltips = TOOLTIPS
hover.mode = 'mouse'

p = gridplot([[select, interact_plot]])
```
